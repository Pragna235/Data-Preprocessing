# 5. Encoding Categorical Data

```python
#One Hot Encoder - To transform categorical data into numerical data

cat_features = ['sex', 'on_thyroxine', 'query_on_thyroxine',
       'on_antithyroid_meds', 'sick', 'pregnant', 'thyroid_surgery',
       'I131_treatment', 'query_hypothyroid', 'query_hyperthyroid', 'lithium',
       'goitre', 'tumor', 'hypopituitary', 'psych', 'TSH_measured',
       'T3_measured', 'TT4_measured', 'T4U_measured',
       'FTI_measured', 'TBG_measured',  'referral_source']

df_one_hot_encoded = pd.get_dummies(data,columns=cat_features,drop_first=True)

df_one_hot_encoded
```

```python
# Get the column names after one hot encoding
df_one_hot_encoded.columns
```

```python
#Label Encoder
label = LabelEncoder()
df_one_hot_encoded['target'] = label.fit_transform(df_one_hot_encoded['target'])
df_one_hot_encoded
```

```python
df_one_hot_encoded.columns
```

```python
#Data transformation using KNN Imputer

imputer = KNNImputer(n_neighbors=2)
data_filled = imputer.fit_transform(df_one_hot_encoded)
data_filled = pd.DataFrame(data_filled,columns = ['age', 'TSH', 'T3', 'TT4', 'T4U', 'FTI', 'TBG', 'target', 'patient_id',
       'sex_M', 'on_thyroxine_t', 'query_on_thyroxine_t',
       'on_antithyroid_meds_t', 'sick_t', 'pregnant_t', 'thyroid_surgery_t',
       'I131_treatment_t', 'query_hypothyroid_t', 'query_hyperthyroid_t',
       'lithium_t', 'goitre_t', 'tumor_t', 'hypopituitary_t', 'psych_t',
       'TSH_measured_t', 'T3_measured_t', 'TT4_measured_t', 'T4U_measured_t',
       'FTI_measured_t', 'TBG_measured_t', 'referral_source_SVHC',
       'referral_source_SVHD', 'referral_source_SVI', 'referral_source_WEST',
       'referral_source_other'])
data_filled
```

```python
#Verify if all the missing values are dealt with
data_filled.isnull().sum()
```

```python
# reindex the output attribute for east data splitting
new_cols = ['age', 'TSH', 'T3', 'TT4', 'T4U', 'FTI', 'TBG', 'patient_id',
       'sex_M', 'on_thyroxine_t', 'query_on_thyroxine_t',
       'on_antithyroid_meds_t', 'sick_t', 'pregnant_t', 'thyroid_surgery_t',
       'I131_treatment_t', 'query_hypothyroid_t', 'query_hyperthyroid_t',
       'lithium_t', 'goitre_t', 'tumor_t', 'hypopituitary_t', 'psych_t',
       'TSH_measured_t', 'T3_measured_t', 'TT4_measured_t', 'T4U_measured_t',
       'FTI_measured_t', 'TBG_measured_t', 'referral_source_SVHC',
       'referral_source_SVHD', 'referral_source_SVI', 'referral_source_WEST',
       'referral_source_other','target']
data_filled = data_filled.reindex(columns=new_cols)
data_filled
```

# 6. Splitting dataset into training and test sets

```python
#train and test splitting
x = data_filled.iloc[:,:-1].values
y = data_filled.iloc[:,-1].values

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1,random_state=0)

x_train
```

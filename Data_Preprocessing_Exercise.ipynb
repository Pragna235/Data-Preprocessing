# 5. Encoding Categorical Data

```python
#One Hot Encoder - To transform categorical data into numerical data

cat_features = ['Gender','Age Group','Cholesterol Level','Blood Pressure','Exercise','Smoker']

df_one_hot_encoded = pd.get_dummies(data,columns=cat_features,drop_first=True)

df_one_hot_encoded
```

```python
#Label Encoder
label = LabelEncoder()
df_one_hot_encoded['Disease Status'] = label.fit_transform(df_one_hot_encoded['Disease Status'])
df_one_hot_encoded
```

```python
#Data transformation using KNN Imputer

imputer = KNNImputer(n_neighbors=2)
data_filled = imputer.fit_transform(df_one_hot_encoded)
data_filled = pd.DataFrame(data_filled,columns = ['BMI','Blood Sugar Level','Disease Status','Gender_Male','Age Group_Senior','Age Group_Young','Cholesterol Level_Normal','Blood Pressure_Low','Blood Pressure_Normal','Exercise_Yes','Smoker_Yes'])
data_filled
```

```python
# reindex the output attribute for east data splitting
new_cols = ['BMI','Blood Sugar Level','Gender_Male','Age Group_Senior','Age Group_Young','Cholesterol Level_Normal','Blood Pressure_Low','Blood Pressure_Normal','Exercise_Yes','Smoker_Yes','Disease Status']
data_filled = data_filled.reindex(columns=new_cols)
data_filled
```

# 6. Splitting dataset into training and test sets

```python
#train and test splitting
x = data_filled.iloc[:,:-1].values
y = data_filled.iloc[:,-1].values

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1,random_state=0)

x_train
```

# 7. Feature Scaling


```python
# Standardization or Normalization of data
# Standardizing

stdsc = StandardScaler()
x_train = stdsc.fit_transform(x_train)
x_train
```
